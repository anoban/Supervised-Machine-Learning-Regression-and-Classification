```python
fig, axes = plt.subplots()
fig.set_size_inches(20, 10)
plt.rcParams["font.size"] = 16

weights = np.array([3.645, 2.55, 1.45015, 0.03265, 0.000045], dtype = np.float32)
x = np.arange(start = 100, stop = 400, step = 2.5)

for i in range(weights.size):
    y = np.zeros(shape = x.size)
    for j in range(i):
        y += weights[i] * np.power(x, j)
    axes.scatter(x, y, color = "green", alpha = 0.5, edgecolor = "black", label = '$' + "".join([f"w_{j+1}x^{j+1} + " for j in range(i)]) + '$')
axes.legend()        
axes.set_ylabel(r"$Sales~price~(\$)$")
axes.set_xlabel(r"$Area~(m^2)$")
plt.show()
```
