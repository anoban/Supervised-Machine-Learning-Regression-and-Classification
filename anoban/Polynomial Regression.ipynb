# ___Polynomial Regression___
-------------------

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# let's say that we have a house sales dataset

area = np.arange(start = 200, stop = 1000)
price = np.square(area) + np.power(area, 3) + np.power(area, 4) - np.random.normal(loc = 0, scale = 100000, size = area.size)
```

```python
fig, axes = plt.subplots()
axes.scatter(area, price, color = "red", alpha = 0.5, edgecolor = "black")
plt.show()
```
