```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# here's the conceptual view of the gradient descent algorithm,
# since our cost function is dependent on two parameters w and b, change of the cost function with unit change in one parameter gives us only a 
# partial derivative not the full derivative.
# hence the use of dabba instead of the usual d to denote the derivative
```

### __repeat until convergence__ $\rightarrow$
### $w = w - \alpha\frac{\partial}{\partial{w}}j(w, b)$        
### $b = b - \alpha\frac{\partial}{\partial{b}}j(w, b)$

```python
# alpha is the learning rate, which dictates how large the stride taken by the gradient descent are!
# the terms on the right are called partial derivatives.
# partial derivatives are typically represented using the "dabba" sign while derivatives are represented using the alphabet "d"
```

## $\mathrm{d}x - derivative$
## $\partial{x} - partial~derivative$

```python
# the learning rate controls how big the leap taken by gradient descent is.
```

## ___Univariate weight-only gradient descent___
-------------------

```python
# imagine a cost function that depends only on the weight, no bias is involved.
# the linear regression model will be in the form of y = wx and the cost function j(w), 
# so our update equation becomes,
# now we are dealing with a full derivative not a partial derivative since the cost function now depends only on the weight!!!
```

# $w = w - \alpha\frac{\mathrm{d}}{\mathrm{dw}}j(w)$

```python
cost = lambda preds, targs: np.square(preds - targs).sum() / (preds.size * 2)
x = np.random.randint(low = 0, high = 100, size = 1000)
y = 0.5647687 * x   # that's the target variable 
w = np.linspace(start = -10, stop = 10, num = 100)
e = np.array(tuple(cost(x * _, y) for _ in w))
```

```python
# since we do not have a bias term b, our plot will be a two dimensional one not a 3D surface plot!

fig, axes = plt.subplots()
axes.plot(w, e)
axes.grid()
axes.set_xlabel("$w$")
axes.set_ylabel("$j(w)$")
plt.show()
```

# $\frac{\mathrm{d}}{\mathrm{dw}}j(w) \rightarrow$

```python
# what does the above term mean?
# the derivative of cost function j(w) for a given value of w is the slope of the graph at the given w value.

fig, axes = plt.subplots()
axes.plot(w, e)
axes.plot(w[9], e[9], mec = "black", ms = 6, mfc = "orange", marker = 'o', linestyle = "none", label = "p1")
axes.plot(w[90], e[90], mec = "black", ms = 6, mfc = "yellow", marker = 'o', linestyle = "none", label = "p2")
axes.grid()
axes.set_xlabel("$w$")
axes.set_ylabel("$j(w)$")
axes.legend()
plt.show()
```

```python
# derivative of j(w) with respect to w is the slope of the curve at those given values for w
# for the point on the right (p2),

print(f"w = {w[90]} and cost = {e[90]}")
```

```python
# and that point has a positive slope! i.e the angle of the x axis with the tangent line is < 90 degrees

# when we update w in gradient descent using w -= alpha * derivative, we end up subtracting a positive value from w
# moving right to left along the x axis, gradient descent lowers the w
```

```python
fig, axes = plt.subplots()
axes.plot(w, e)
axes.plot(w[9], e[9], mec = "black", ms = 6, mfc = "orange", marker = 'o', linestyle = "none", label = "p1")
axes.plot(w[90], e[90], mec = "black", ms = 6, mfc = "yellow", marker = 'o', linestyle = "none", label = "p2")
axes.fill_betweenx(y = np.arange(axes.get_xlim()[0], axes.get_ylim()[1], step = 100), x1 = w[np.argmin(e)], x2 = w.max(), color = "red", alpha = 0.1)
axes.grid()
axes.set_xlabel("$w$")
axes.set_ylabel("$j(w)$")
axes.legend()
plt.show()
```

```python
# as we move right -> left in the red region, the curve reaches a minimum for cost, approximately at w = 0.5 
# and this is eaxctly what we want, to minimize the cost function

w[np.argmin(e)]
```

```python
# let's move from left to right, starting from the point p1

fig, axes = plt.subplots()
axes.plot(w, e)
axes.plot(w[9], e[9], mec = "black", ms = 6, mfc = "orange", marker = 'o', linestyle = "none", label = "p1")
axes.plot(w[90], e[90], mec = "black", ms = 6, mfc = "yellow", marker = 'o', linestyle = "none", label = "p2")
axes.fill_betweenx(y = np.arange(axes.get_xlim()[0], axes.get_ylim()[1], step = 100), x2 = w.min(), x1 = w[np.argmin(e)], color = "blue", alpha = 0.1)
axes.grid()
axes.set_xlabel("$w$")
axes.set_ylabel("$j(w)$")
axes.legend()
plt.show()
```

```python
# on the left side, the slope is negative, i.e the angle the tangent lines makes with the x axis is > 90 degrees.
# this point will consequently have a negative slope.

# when we update the weight in gradient descent, using w -= alpha * derivative, we'll be adding a positive term to the w
# which will help us move towards right, where the minimum for the cost function is located!
```
