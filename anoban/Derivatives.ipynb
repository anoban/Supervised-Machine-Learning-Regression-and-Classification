```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# here's the conceptual view of the gradient descent algorithm,
```

## _repeat until convergence {_
## $$w = w - \alpha\frac{\partial}{\partial{w}}j(w, b)$$
## $$b = b - \alpha\frac{\partial}{\partial{b}}j(w, b)$$
## _}_

```python
# alpha is the learning rate, which dictates how large the stride taken by the gradient descent are!
# the terms on the right are called partial derivatives.
# partial derivatives are typically represented using the "dabba" sign while derivatives are represented using the alphabet "d"
```

## $\mathrm{d}x - derivative$
## $\partial{x} - partial~derivative$

```python
# the learning rate controls how big the leap taken by gradient descent is.
```

## ___Univariate weight only gradient descent___
-------------------

```python
# imagine a cost function that depends only on the weight, no bias involved.
# the linear regression model will be in the form of y = wx
# and the cost function j(w), so our update equation becomes,
```

# $w = w - \alpha\frac{\partial}{\partial{w}}j(w)$

```python
cost = lambda preds, targs: np.square(preds - targs).sum() / (preds.size * 2)
x = np.random.randint(low = 0, high = 100, size = 1000)
y = 0.5647687 * x   # that's the target variable 
w = np.linspace(start = -10, stop = 10, num = 100)
e = np.array(tuple(cost(x * _, y) for _ in w))
```

```python
# since we do not have a bias term b, our plot will be a two dimensional one not a 3D surface plot!

fig, axes = plt.subplots()
axes.plot(w, e)
axes.grid()
axes.set_xlabel("$w$")
axes.set_ylabel("$j(w)$")
plt.show()
```

# $\frac{\partial}{\partial{w}}j(w) \rightarrow$

```python
# what does the above term mean?
# the partial derivative of w for a given value with respect to the cost function j(w) is the slope of the graph at the given w value.


fig, axes = plt.subplots()
axes.plot(w, e)
axes.plot([w[9], w[90]], [e[9], e[90]], mec = "black", ms = 5, mfc = "red", marker = 'o', linestyle = "none")
axes.grid()
axes.set_xlabel("$w$")
axes.set_ylabel("$j(w)$")
plt.show()
```

```python
# partial derivative of j(w) with respect to w is the slope of the curve at those given values for w
# for the point on the right,

print(f"w = {w[90]} and cost = {e[90]}")
```

```python
# and that point has a positive slope!
# when we update w in gradient descent, 
# w -= alpha * pderiv, we end up subtracting a positive value from w
# so the update decreases the w, moving right - left along the x axis
```
