# ___Overfitting___
-------------------

```python
import numpy as np
np.random.seed(2024 - 5 - 26)
import matplotlib.pyplot as plt
```

## ___Regression___
-------------------------

```python
# consider the follwong synthetic data

area = np.linspace(start = 10, stop = 250, num = 200)
prices = area * (2.56465 + np.random.rand(area.size))
```

```python
fig, axes = plt.subplots()
fig.set_size_inches(6, 6)
axes.scatter(area, prices, color = "white", edgecolor = "red")
axes.grid()
axes.set_xlabel("area")
axes.set_ylabel("price")
plt.show()
```

```python
# to the above distribution, we could fit a straight line using linear regression!
```

```python
fig, axes = plt.subplots()
fig.set_size_inches(6, 6)
axes.scatter(area, prices, color = "white", edgecolor = "blue")
axes.plot(area, area * 3, linestyle = "-.", color = "red")
axes.grid()
axes.set_xlabel("area")
axes.set_ylabel("price")
plt.show()
```

```python
# this is a descent fitting model
```

```python
area = np.linspace(start = 10, stop = 250, num = 200)
# using polynomial features
prices = (area * 1.7853) + (0.005342 * np.power(area, 2)) + (0.0000456 * np.power(area, 3)) + \
            np.random.normal(loc = 0, scale = 100, size = area.size) # introducing some random noise
prices_denoised = (area * 1.7853) + (0.005342 * np.power(area, 2)) + (0.0000456 * np.power(area, 3))
fig, axes = plt.subplots()
fig.set_size_inches(20, 8)
axes.scatter(area, prices, color = "white", edgecolor = "blue")
axes.plot(area, area * 4.7893, linewidth = 1.5, color = "red", label = "under fit")
axes.plot(area, prices_denoised, linewidth = 1.5, color = "black", label = "good fit")
axes.plot(area, prices, linewidth = 1.5, color = "green", label = "over fit")
axes.legend()
axes.set_xlabel("area")
axes.set_ylabel("price")
plt.show()
```

```python
# that's an amazing plot :)
```

```python
# the red mode is underfitting the data, i.e it fails to accurately capture the relationships between the features and the target
# such models are considered to have high bias!
# another way to imagine this is to see the model as a highly prejudiced one, no matter how complex the relationships are between the features and
# the target, this model always opts for an oversimplified linear relationship!
# hence, underfits the data
```

```python
# the black model is a good enough model that fits the data reasonably well :)
# this is a model that could generalize well and give decent predictions for previously unknown data!
# a phenomenon known as generalization!
```

```python
# the green model, on the other hand, is overfitting the data.
# this model fits all the datapoint in the training set precisely!
# the cost function will be 0.0000 exactly! and this model will show a 100% accuracy if tested against the training data
# but will perform very poorly on new data because it has become so specialized that it has learned intricate nuances in the training data
# when this model tries to replicate these very surgical relationships with test data, the predictions may become very deviant if the test data
# does not come from a very very similar context!

# such models are considered to have high variance!
```

## ___Classification___
------------------

```python
# underfitting and overfitting applies to classification problems too.
# when a classifier has underfitted the training data, it has learned a very vague decision boundary which could lead the model to misclassify 
# many data points.
# in this case, the classifier has failed to identify a boundary that reasonably delimits the target categories 
```

```python
# in overfitting the decision boundary becomes extremely jagged and specific to the training data
# hence the model will perform incredibly well on training data but may fail miserably in generalizing to new test data!
```
